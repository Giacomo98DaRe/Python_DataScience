```python
import numpy as np
import pandas as pd
```

##### FROM AN ARRAY

```python
# Create an array
first_data_example = np.array([[1, 2], [2, 4]])
first_data_example
```

```python
# Common name for a DataFrame: df

first_dataframe = pd.DataFrame(first_data_example, index=["lol", "lel"], columns=["col1", "col2"])
first_dataframe
```
