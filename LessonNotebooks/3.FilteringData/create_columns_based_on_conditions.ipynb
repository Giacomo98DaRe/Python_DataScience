```python
import pandas as pd
import numpy as np

from utilities.reader_tools.GenericCSVReader import DataframeFromCSV
```

##### BASED ON 1 CONDITION

```python
laptop_df = DataframeFromCSV("laptop_price.csv", encoding="cp1252")
```

```python
laptop_df.head(3)
```

```python
price_tier = np.where(laptop_df["Price_euros"] > 2000, "Expensive", "Cheap")
```

```python
laptop_df["Price_tier"] = price_tier
laptop_df.head(5)
```

```python
laptop_df["Price_tier"].value_counts()
```

##### EXERCISE

If screen size is > 15 inches -> screen is big

```python
screen_dimension = np.where(laptop_df["Inches"] > 15, "Big", "Small")
screen_dimension
```

```python
laptop_df["Screen Dimension"] = screen_dimension
```

```python
laptop_df.head(5)
```

```python
renamed_df = laptop_df.rename(columns={"Screen Dimension" : "Screen_Dimension"})
```

```python
renamed_df.head(5)
```

```python
renamed_df["Screen_Dimension"].value_counts()
```
