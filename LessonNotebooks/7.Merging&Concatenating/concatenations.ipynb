##### CONCATENATE VERTICALLY 

Immagino di avere due df con delle colonne in comune. La concatenazione verticale prevede porre i due df uno sotto l'altro e semplicemente "incollarli". Di default, i nomi degli indici non verranno controllati -> potrò avere doppio indexes. Problema risolvibile con parametro ignore indexes.

```python
ex1 = pd.DataFrame({"id": [1, 2],
                    "age":[29, 30]})

ex2 = pd.DataFrame({"id": [3, 4],
                    "age":[39, 40]})
```

```python
pd.concat([ex1, ex2], ignore_index=True)
```

##### CONCATENATE HORIZONTALLY

We need the same indexes if we want to perform this concatenation.

```python
ex1 = pd.DataFrame({"id": [1, 2],
                    "age":[29, 30]})

ex2 = pd.DataFrame({"job": ["Doctor", "Plumber"]})
```

```python
ex2
```

```python
pd.concat([ex1, ex2], axis=1)
```
