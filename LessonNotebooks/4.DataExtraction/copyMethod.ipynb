```python
import pandas as pd
from utilities.reader_tools import GenericDatasetReader

df_players = GenericDatasetReader.DataframeFromCSV("players_20.csv", setIndexName="short_name")
df_players.head()
```

##### COPY METHOD

```python
# Deep = True
df_copy = df_players.copy()
```

```python
df_players.loc["L. Messi", "height_cm"] = 190
df_players.head()
```

```python
df_copy.head()
```

```python
# Deep False -> make dataframe dipendent. Changes will be reflected

df_shallow_copy = df_copy.copy(deep=False)
```

```python
df_copy.loc["Cristiano Ronaldo", "height_cm"] = 210
df_copy.head()
```

```python
df_shallow_copy.head()
```

```python
# Simple assignment -> it is a shallow copy

df_new_copy = df_players
```
