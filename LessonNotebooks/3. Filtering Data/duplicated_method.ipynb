##### EXAMPLE 2

Find cheapest and most expensive laptop

```python
sorted_df = df_laptops.sort_values(["Company", "Price_euros"])
sorted_df
```

```python
sorted_df.value_counts("Company")
```

```python
duplicated_first = sorted_df.duplicated("Company")
```

```python
sorted_df[~duplicated_first][["Company", "Price_euros"]]
```

```python
last_filter = sorted_df.duplicated(["Company"], keep="last")
```

```python
sorted_df[~last_filter][["Company", "Price_euros"]]
```

```python
# With keep = false, you keep neither the first nor the last
duplicated_keepFalse = sorted_df.duplicated(["Company"], keep=False)
duplicated_keepFalse
```

```python
# Every item has at least one duplicate of company
sorted_df[~duplicated_keepFalse]
```

##### EXERCISE

Find the biggest and smallest screen size

###### DUPLICATES

```python
df_laptops.head(3)
```

```python
sorted_by_size = df_laptops.sort_values(["Inches", "Company"])
sorted_by_size
```

```python
filter_minimum = sorted_by_size.duplicated(["Company"])
filter_minimum
```

```python
min_list = sorted_by_size[~filter_minimum][["Company", "Inches"]]
min_list
```

```python
filter_max = sorted_by_size.duplicated(["Company"], keep="last")
filter_max
```

```python
sorted_by_size[~filter_max][["Company", "Inches"]]
```

###### DROP DUPLICATES

```python
sorted_by_size.head(3)
```

```python
drop_min = sorted_by_size.drop_duplicates(["Company"])[["Company", "Inches"]]
```

```python
min_list.equals(drop_min)
```

```python
drop_max = sorted_by_size.drop_duplicates(["Company"], keep="last")
drop_max[["Company", "Inches"]]
```
