```python
import pandas as pd
from utilities.reader_tools import GenericCSVReader

df_players = GenericCSVReader.DataframeFromCSV("players_20.csv", setIndexName="short_name")
df_players.head()
```

##### QUERY

Filtering dataset without boolean slicing (the previous method)

```python
# Query
# df_players.query("age>34")

# Boolean
age_condition = df_players["age"] > 34
df_players.loc[age_condition, "age"]
```

```python
# Multiple
df_players.query("age>34 & nationality=='Italy'")
```

```python
# Convert height to meters and select the ones above 1.8
# df_players.query("height_cm/100 > 1.8")

height_condition = (df_players["height_cm"] / 100) > 1.8
df_players[height_condition]
```

```python
# Players born after 1990

# Convert datetime type
# df_players["dob"] = df_players["dob"].astype("datetime64[s]")
# Better alternative:
df_players["dob"] = pd.to_datetime(df_players["dob"])
```

```python
# Query
df_players["dob"].dt.year  # type: ignore

filtered_fromData = df_players.query("dob.dt.year > 1990")
filtered_fromData["dob"]
```
