```python
import pandas as pd
import numpy as np
from utilities.reader_tools.GenericDatasetReader import DataframeFromCSV

raw_population_df = DataframeFromCSV("population_total.csv")
raw_population_df
```

##### BOXPLOT

```python
raw_population_df.dropna(inplace=True)
population_df = raw_population_df.pivot(index="year", columns="country", values="population")
population_df
```

###### SINGLE BOXPLOT

```python
population_df["United States"].plot(kind="box", color="green", ylabel="Population")
```

###### MULTIPLE BOXPLOT

```python
population_df = population_df[["United States", "China", "Indonesia", "Brazil"]]
population_df.plot(kind="box", color="green", ylabel="Population")
```

##### HISTOGRAM

```python
population_df[["Indonesia"]]
population_df[["Indonesia"]].plot(kind="hist", title="Indonesia population x year")
```

##### SCATTER PLOT

```python
condition = raw_population_df["country"].isin(["United States", "China", "Brazil"])
scatter_ex = raw_population_df[condition]
scatter_ex
```

```python
scatter_ex.plot(kind="scatter", 
                x="year",
                y="population",
                s=3,
                color="red")
```
