##### DICTIONARY OF CSV

```python
root = "https://www.football-data.co.uk/mmz4281/"
serieA_suffix = "/I1.csv"
premierLeague_suffix = "/E0.csv"
```

```python
leagues = {"Premier":premierLeague_suffix, "SerieA":serieA_suffix}
seasons = ["2425", "2526"]

dict_leagues = {}

for league, suffix in leagues.items():
    selected_league = []
    for season in seasons:
        link = root + season + suffix
        print(link)
        league_df = pd.read_csv(link)
        selected_league.append(league_df)
    
    dict_leagues[league] = pd.concat(selected_league)
```

```python
for league, tabel in dict_leagues.items():
    print(len(tabel))
```
